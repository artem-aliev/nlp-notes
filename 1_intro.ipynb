```python
def mse (a, b):
  l = len(a)
  sum=0
  for i, j in zip(a,b):
    sum+=(i-j)**2
  return sum/l
```

```python
def mse2 (a,b):
  return np.sum( (a-b)**2) / len(a)
mse2
```

```python
def predict(a,b, x):
  return a*(x**.5) + b

def to_min (a, b):
  yp = predict(a,b,xs)
  return mse2(ys, yp)
```

```python
sa = np.arange(-2, 2, 0.1)
sy= np.array([to_min(a, 0) for a in sa])
plt.plot(sa, sy)
```
